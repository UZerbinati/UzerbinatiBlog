Now our aim is to define a function that given an image create an histogram on the feature space, in this particular case the feature space is gona be the 1D color space. Later we will implement a funtion that does the same in the 3D color space in particular in the CIE-LAB feature space.

```python
def GSHistogram(Image,N):
        H = [];
        step = 255/N;
        mask = np.logical_and(Image >= 0, Image <= step);
        Pixels = (Image[mask]).size;
        H.append(Pixels);
        for i in range(1,N):
            mask = np.logical_and(Image > step*i, Image <= step*(i+1));
            Pixels = (Image[mask]).size;
            H.append(Pixels);
        assert(sum(H) == Image.size);
        return H;
H = GSHistogram(camera,5)
H
```

```python
def LABHistogram(Image,N,M):
        H = [];
        step = 100/N;
        mask = np.logical_and(Image[:,:,0] >= 0, Image[:,:,0] <= step);
        step2 = 255/M;
        mask2 = np.logical_and(Image[mask,1] >= -128, Image[mask,1] <= -128+step2);
        mask3 = np.logical_and(Image[mask][mask2,2] >= -128, Image[mask][mask2,2] <= -128+step2);
        Pixels = (Image[mask][mask2][mask3,2]).size;
        H.append(Pixels);
        for j in range(1,M):
                mask3 = np.logical_and(Image[mask][mask2,1] > -128+step2*j, Image[mask][mask2,1] <= -128+step2*(j+1));
                Pixels = (Image[mask][mask2][mask3,2]).size;
                H.append(Pixels);
        for i in range(1,M):
            mask2 = np.logical_and(Image[mask,1] > -128+step2*i, Image[mask,1] <= -128+step2*(i+1));
            mask3 = np.logical_and(Image[mask][mask2,2] >= -128, Image[mask][mask2,2] <= -128+step2);
            Pixels = (Image[mask][mask2][mask3,2]).size;
            H.append(Pixels);
            for j in range(1,M):
                mask3 = np.logical_and(Image[mask][mask2,1] > -128+step2*j, Image[mask][mask2,1] <= -128+step2*(j+1));
                Pixels = (Image[mask][mask2][mask3,2]).size;
                H.append(Pixels);
                
        for n in range(1,N):
            mask = np.logical_and(Image[:,:,0] > step*n, Image[:,:,0] <= step*(n+1));
            mask2 = np.logical_and(Image[mask,1] >= -128, Image[mask,1] <= -128+step2);
            mask3 = np.logical_and(Image[mask][mask2,2] >= -128, Image[mask][mask2,2] <= -128+step2);
            Pixels = (Image[mask][mask2][mask3,2]).size;
            H.append(Pixels);
            for j in range(1,M):
                mask3 = np.logical_and(Image[mask][mask2,1] > -128+step2*j, Image[mask][mask2,1] <= -128+step2*(j+1));
                Pixels = (Image[mask][mask2][mask3,2]).size;
                H.append(Pixels);
            for i in range(1,M):
                mask2 = np.logical_and(Image[mask,1] > -128+step2*i, Image[mask,1] <= -128+step2*(i+1));
                mask3 = np.logical_and(Image[mask][mask2,2] >= -128, Image[mask][mask2,2] <= -128+step2);
                Pixels = (Image[mask][mask2][mask3,2]).size;
                H.append(Pixels);
                for j in range(1,M):
                    mask3 = np.logical_and(Image[mask][mask2,1] > -128+step2*j, Image[mask][mask2,1] <= -128+step2*(j+1));
                    Pixels = (Image[mask][mask2][mask3,2]).size;
                    H.append(Pixels);
        assert(sum(H)==Image[:,:,1].size)
        assert(len(H)==N*M*M)
        return H;
    
def AverageHistogram(H,tol):
    AverageH = [];
    for i in range(0,len(H)):
        if H[i] > tol:
            AverageH.append(H[i])
    return AverageH;

H = LABHistogram(LABCat,4,8)
AH = AverageHistogram(H,LABCat[:,:,0].size/1000);
```

Now our aim will be the one of introduing some bin to bin distance to compare the LAB histogram that we have obtained using the function preaviusly defined. The first distance we introduce is the Minkowski distance wich is an extension of the $L^p$ distance for histograms, that is defined as:
$$ d_r=\bigg( \sum_i |h_i-k_i|^r \bigg)^\frac{1}{r}$$

```python
def MinkowskiHDistance(H,K,r):
    assert((len(H)==len(K)));
    S = 0;
    for i in range(0,len(H)):
        S = S+abs(H[i]-K[i])**r;
    S = S**(1/r);
    return S;
```

```python
MinkowskiHDistance(H,H,1)
```
